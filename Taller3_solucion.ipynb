**1.1 Lectura de archivo:**

```python
with open('UNA_SOLITARIA_VOZ_HUMANA.txt', encoding='utf-8') as f:
    for line in f:
        print(str(line))
    f.close()
```

```python
with open("UNA_SOLITARIA_VOZ_HUMANA.txt", encoding="utf-8") as t1:
    solitaria = t1.read()
    t1.close()
```

Remover y reemplazar caracteres:

```python
solitaria = solitaria.replace('\xa0', ' ').replace('\n',' ')
```

**1.2 Convertir a minúsculas:**

```python
solitaria_lower = solitaria.lower()
#solitaria_lower
```

**1.3 Conservar palabras:**

```python
solitaria_w = re.sub(r"[\W\d_]+", " ", solitaria_lower)
#solitaria_w
```

Por conveniencia, se van a quitar las tildes, para ello se define la siguiente función:

```python
def quitar_tildes(s):
    tildes = (
        ("á", "a"),
        ("é", "e"),
        ("í", "i"),
        ("ó", "o"),
        ("ú", "u"),
    )
    for origen, destino in tildes:
        s = s.replace(origen, destino)
    return s
```

```python
solitaria_sin_tildes = quitar_tildes(solitaria_w)
solitaria_sin_tildes
```

**1.4 Tokenizar:**

```python
solitaria_t = solitaria_sin_tildes.split()
solitaria_t
```

# Punto 2:

- `[8pts]` Quitar palabras vacias
- `[9pts]` ¿Cuáles son las 10 palabras no vacias más usadas?

**2.1 Quitar palabras vacias:**

```python
nltk.download('stopwords')

stopwords_sp = stopwords.words('spanish')
```

```python
solitaria_l = [palabra for palabra in solitaria_t if palabra not in stopwords_sp]
solitaria_l
```

**2.2 Diez palabras mas comunes:**

```python
Counter(solitaria_l).most_common(10)
```

# Punto 3:

- `[8pts]` Stemming del documento
- `[9pts]` ¿Cuáles son las 10 raíces más usadas?

**3.1 Stemming del documento:**

```python
spanishStemmer = SnowballStemmer("spanish") # carga de raices en español

solitaria_stem = []

for palabra in solitaria_l:
    raiz = spanishStemmer.stem(palabra)
    solitaria_stem.append(raiz)
    
solitaria_stem
```

**3.2 Diez raices mas usadas:**

```python
Counter(solitaria_stem).most_common(10)
```
