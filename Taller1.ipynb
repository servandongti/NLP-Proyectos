### 3. Contar número de palabras por estudiante

```python
# para cada comentario del estudiante se cuenta el numero de palabras

conteo = []

for i in range(0,data.shape[0]):
    conteo.append(len(data['¿Por qué estás estudiando análitica estratégica de datos?'][i].split()))
```

```python
# unir el conteo de palabras al dataframe

data = pd.concat([data, pd.DataFrame(conteo, columns=['conteo_palabras'])], axis=1)
data
```

### 4.1 Palabra más utilizada por estudiante

**Definición de funciones:**

```python
def palabra_mas_usada(s):
    lista_palabras = s.split()
    
    frecuencia_palabras = []

    for i in lista_palabras:
        frecuencia_palabras.append(lista_palabras.count(i))
        
    max_w = max(frecuencia_palabras)
    posicion_max = frecuencia_palabras.index(max_w)
    
    return lista_palabras[posicion_max]
```

```python
def frec_palabra_mas_usada(s):
    lista_palabras = s.split()
    
    frecuencia_palabras = []

    for i in lista_palabras:
        frecuencia_palabras.append(lista_palabras.count(i))
        
    return max(frecuencia_palabras)
```

```python
# para cada comentario del estudiante se calcula la palabra mas frecuente

palabra_frec_estudiante = []

for i in range(0,data.shape[0]):
    palabra_frec_estudiante.append(palabra_mas_usada(data['¿Por qué estás estudiando análitica estratégica de datos?'][i]))
```

```python
# para cada comentario del estudiante se calcula la frecuencia maxima por palabra

frec_palabra_estudiante = []

for i in range(0,data.shape[0]):
    frec_palabra_estudiante.append(frec_palabra_mas_usada(data['¿Por qué estás estudiando análitica estratégica de datos?'][i]))
```

```python
# unir la palabra frecuente con su frecuencia al dataframe

pd.concat([data, pd.DataFrame(palabra_frec_estudiante, columns=['palabra_frecuente']),
          pd.DataFrame(frec_palabra_estudiante, columns=['frecuencia_palabra'])], axis=1)
```
